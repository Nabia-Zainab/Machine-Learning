```python
import pandas as pd
import  requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
```

```python
# Convert coming data into JSON format
response.json()
```

```python
response.json()['results']
```

```python
# Convert into dataframes
pd.DataFrame(response.json()['results']).head(2)
```

```python
# Extract the feactures I need
pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
# We use loop, loop will run according to number of pages.
# At every loop, we will create a dataframe and append this dataframe into a old dataframe

# Make a empty dataframe
df = pd.DataFrame()
df
```

```python
for i in range(1,429):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
  temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
  df = df.append(temp_df, ignore_index=True) # LOOP-1 --> index 1-20 LOOP-2 --> index 1-20 ...
```

```python
df.to_csv('movie.csv')
```
