```python
data = []

for i in range(1, 10):
    webpages = requests.get("https://www.dawn.com/news/{}".format(i)).text
    soup = BeautifulSoup(webpages, 'lxml')

    title = soup.find('a', class_='story__link').text
    publish = soup.find('span', class_='timestamp--date').text.strip()
    paragraph = soup.find('div', class_='story__content').text.strip()

    data.append({'title': title, 'publish': publish, 'paragraph': paragraph})

final = pd.DataFrame(data)
# print(final)

final.to_csv('dawn_news.csv', index=False)
```
